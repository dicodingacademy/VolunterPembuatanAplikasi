# Algoritma Machine Learning K-Nearest Neighbors (KNN) untuk klasifikasi penyakit jantung

Algoritma KNN ini dibuat tanpa menggunakan bantuan library, murni hasil mikir sendiri untuk memenuhi tugas mata kuliah intelligent system ðŸ˜‰.

1. Dataset (1309):  
https://www.kaggle.com/datasets/bharath011/heart-disease-classification-dataset
2. Fitur (8):  
age, gender, impluse, pressurehight, pressurelow, glucose, kcm, troponin
3. Class/Label (2):  
positive, negative

```python
import pandas as pd
import math
```

```python
# Membaca Dataset

datasetName = 'heart_attack.csv'
dataset = pd.read_csv(datasetName)
dataset.info()
```

```python
# Menghapus data yang semua nilainya 0

dataset = dataset.loc[(dataset!=0).any(axis=1)]
dataset.info()
```

```python
# Menginput data testing

features_all = ['age', 'gender', 'impluse', 'pressurehight', 'pressurelow', 'glucose', 'kcm', 'troponin', 'class']
features = ['age', 'gender', 'impluse', 'pressurehight', 'pressurelow', 'glucose', 'kcm', 'troponin']

test_values = [
    [71, 1, 71, 119, 76, 159, 0.468, 0.029, 'positive'],
    [53, 1, 73, 135, 81, 115, 165.1, 0.014, 'positive'],
    [43, 0, 68, 116, 74, 81, 1.64, 0.015, 'positive'],
    [66, 0, 70, 113, 62, 266, 300, 0.012, 'positive'],
    [67, 1, 87, 148, 89, 142, 1.87, 0.01, 'negative'],
    [51, 0, 85, 140, 82, 101, 1.69, 0.008, 'negative'],
    [50, 1, 83, 140, 81, 244, 3.27, 2.23, 'positive'],
    [67, 1, 82, 164, 90, 130, 3.75, 0.009, 'negative'],
    [59, 1, 81, 150, 51, 117, 1.51, 1.55, 'positive'],
    [20, 1, 60, 156, 60, 103, 5.22, 1.84, 'positive']
]

test = pd.DataFrame(test_values, columns=features_all)
test
```

```python
# Fungsi untuk menghitung jarak menggunakan rumus Euclidean Distance

def calculate_distance(train, test, column):
  result = 0
  for i in column:
    result = result + pow((train[i] - test[i]), 2)
  result = math.sqrt(result)
  return result
```

```python
# Fungsi untuk menentukan label

def determine_label(train, k, label):
  sorted = train.sort_values(['distance'], ascending=True)
  nearest = sorted.head(k)
  result = {}
  for i in label:
    try:
      result[i] = nearest['class'].value_counts()[i]
    except:
      result[i] = 0
  result['result'] = max(result, key=result.get)
  return result
```

```python
# Menentukan label pada data testing dengan K = 3

label = ['positive', 'negative']
k = 3

for i in range(0, len(test.index)):
  dataset['distance'] = dataset.apply(calculate_distance, axis=1, args=(test.loc[i], features))
  result = determine_label(dataset, k, label)
  test.loc[i, ['positive']] = result['positive']
  test.loc[i, ['negative']] = result['negative']
  test.loc[i, ['prediction']] = result['result']

test
```

```python
# Menentukan kinerja confusion matrix

perf = 'performance'

for i in range(0, len(test.index)):
  real = test._get_value(i, 'class')
  prediction = test._get_value(i, 'prediction')
  if ((real == 'positive') and (prediction == 'positive')):
    test.loc[i, [perf]] = 'TP'
  elif ((real == 'positive') and (prediction == 'negative')):
    test.loc[i, [perf]] = 'FN'
  elif ((real == 'negative') and (prediction == 'negative')):
    test.loc[i, [perf]] = 'TN'
  elif ((real == 'negative') and (prediction == 'positive')):
    test.loc[i, [perf]] = 'FP'

test
```

```python
# Menghitung Presisi, Recall, dan Akurasi pada confusion matrix

try:
  TP = test[perf].value_counts()['TP']
except:
  TP = 0
try:
  TN = test[perf].value_counts()['TN']
except:
  TN = 0
try:
  FP = test[perf].value_counts()['FP']
except:
  FP = 0
try:
  FN = test[perf].value_counts()['FN']
except:
  FN = 0

precision = ((TP) / (FP + TP) * 100) / 100
recall = ((TP) / (FN + TP) * 100) / 100
accuration = ((TP + TN) / (TP + TN + FP + FN) * 100) / 100

print("TP =", TP, "\nTN =", TN, "\nFP =", FP, "\nFN =", FN)
print("Presisi =", precision, "\nRecall =", recall, "\nAkurasi =", accuration)
```
